```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Import CSV file: Obtained from https://www.the-numbers.com/movie/budgets/all/501 

After cleaning the data using excel,it was reduced to 2 columns:

1.Production Budget: Amount spent on the making of the movie or projected amount to be spent

2.Worldwide gross: Revenue obtained from the movie release

```python
data = pandas.read_csv('cost-revenue-dirty.csv')
```

Generate descriptive statistics that summarize the central tendency, dispersion and shape of a datasetâ€™s distribution, excluding NaN values.

```python
data.describe()
```

Creates a dataframe that helps to do exploratory analysis and supports splitting the featurtes and label

```python
X = DataFrame(data, columns=['production_budget'])
y = DataFrame(data, columns=['worldwide_gross'])
```

Plot the scatter plot of the all the points in the data

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)
plt.ylim(0,2783919000.0)
plt.xlim(0,425000000.0)
plt.xlabel('production budget $')
plt.ylabel('worldwide gross $')
plt.title('Production Budget vs Worldwide gross')
```

Learn the model that fits the data best using scikit learn linear regression library

```python
model = LinearRegression()
model.fit(X,y)
```

Get the x coefficient(m) of the line formula y= mx +  c

```python
model.coef_
```

Get the y intercept(c) of the line 

```python
model.intercept_
```

Plot the line that was learnt using linear regression

```python
plt.figure(figsize=(10,6))
plt.scatter(X,y, alpha=0.3)
plt.plot(X,model.predict(X),color='red',linewidth=5)
plt.ylim(0,2783919000.0)
plt.xlim(0,425000000.0)
plt.xlabel('production budget $')
plt.ylabel('worldwide gross $')
plt.title('Production Budget vs Worldwide gross')
```

Get the accuracy of the model

```python
model.score(X,y)
```
